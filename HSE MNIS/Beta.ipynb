```python
import pandas as pd
import yfinance as yf

# Download historical data for all 10 stocks as well as BIST 100 for 2021-2025

tickers = [
    'TUPRS.IS', 
    'TCELL.IS', 
    'CCOLA.IS', 
    'SISE.IS', 
    'ENJSA.IS', 
    'ARCLK.IS', 
    'OTKAR.IS', 
    'SELEC.IS',
    'AKSA.IS',
    'DGNMO.IS',
    'XU100.IS' # BIST 100 Instanbul bourse ticker
]

start_date = "2021-01-01"
end_date = "2025-09-30"
interval = "1d"

# df is a multilevel df with level 0 columns - OHLC and level 1 columns - tickers
df = yf.download(tickers, start=start_date, end=end_date, interval=interval) 
returns = df['Close'].pct_change()
# calculating covariance
covariance = returns.groupby(returns.index.year).apply(lambda g: g.drop(columns=['XU100.IS']).apply(lambda y: y.cov(g['XU100.IS'])))
# calculating variance of stock market index
var_idx = returns.groupby(returns.index.year)['XU100.IS'].var(ddof=1)
# calculating beta
beta = covariance.div(var_idx, axis=0)
beta
```
